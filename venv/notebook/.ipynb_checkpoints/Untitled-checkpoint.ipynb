Find ratio 

HSI index update and show

```python
import pandas as pd
import numpy as np
from StockObj import StockObj
import matplotlib.pyplot as plt
import sqlite3
abc = StockObj("abc", "%5EHSI")
conn = sqlite3.connect("D:\\stks\\tst.db")
cur = conn.cursor()
cur.execute(
    "SELECT close, stkdate FROM STOCKS where idx = " + "'" + "0700.HK" + 
    "'" + "and stkdate between " + "'" + "2016-01-01" + "'" + "and" + "'" + "2019-04-08" + "'")
num = cur.fetchall()
indlist=[]
for t in num:
    if t[0] == '-':
        continue
    #indlist.append(float(t[0].replace(',', '')))
    indlist.append(t[0])
y=indlist

x=np.arange(start = 0, stop = len(indlist), step = 1)

g610x=[]
g610y=[]

g233x=[]
g233y=[]

for t in x:
    if t > 610:
        g610x.append(t)
        g610y.append(np.average(indlist[t-610:t]))
        
for t in x:
    if t > 233:
        g233x.append(t)
        g233y.append(np.average(indlist[t-233:t]))

fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(x, y, color='lightblue', linewidth=3)
ax.scatter(g610x, g610y, color='green', linewidth=1)
ax.scatter(g233x, g233y, color='red', linewidth=1)
plt.savefig('foo.png')
#plt.xticks(range(1, 30))
plt.show()
```
